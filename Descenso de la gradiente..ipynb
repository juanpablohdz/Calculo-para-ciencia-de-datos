```python
from matplotlib import  cm # manejar colores
import numpy as np 
import matplotlib.pyplot as plt
```

GRAFICA 3D DE NUESTRA FUNCION DE COSTE

```python
fig,ax= plt.subplots(subplot_kw={"projection":"3d"}) # grafica 3d
def f(x,y):
    #return x**2 + y**2
    return np.sin(x) + 2*np.cos(y) /4
resolucion=100

x=np.linspace(-4,4,resolucion)
y=np.linspace(-4,4,resolucion)
X,Y=np.meshgrid(x,y)

Z=f(X,Y)

surf=ax.plot_surface(X,Y,Z,cmap=cm.cool) # grafica # colores:https://matplotlib.org/stable/tutorials/colors/colormaps.html
fig.colorbar(surf)# linea de nivel 
```

### DESCENSO DE LA GRADIENTE   
desventaja es que solo llega a minimos locales 

```python
level_map = np.linspace(np.min(Z), np.max(Z),resolucion) #mapas de nivel
plt.contourf(X, Y, Z, levels=level_map,cmap=cm.cool)
plt.colorbar()
plt.title('Descenso del gradiente')
#derviada parcial 
def derivate(_p,p):
  return  (f(_p[0],_p[1]) - f(p[0],p[1])) / h

p = np.random.rand(2) * 8 - 4 # generar dos valores aleatorios punto aleatorio para hacerlo entre -4 a 4 

plt.plot(p[0],p[1],'o', c='k')# PUNTO

lr = 0.01 # lerning rate
h = 0.01#incremento

#DECENDER LA GRADIENTE 
grad = np.zeros(2)

for i in range(10000): #proceso iterativo  actualizar punto
  for idx, val in enumerate(p): 
    cp = np.copy(p)# copia de mi punto 

    cp[idx] = cp[idx] + h;# incremento 

    dp = derivate(cp,p) # calcular la derviada parcial 

    grad[idx] = dp  #guardar componentes de las derviadas

  p = p - lr * grad

  if(i % 10 == 0):# numero de ploteo
    plt.plot(p[0],p[1],'o', c='r')

plt.plot(p[0],p[1],'o', c='w') #punto
plt.show()

print("El punto mnimo se encuentra en: ", p)
```
