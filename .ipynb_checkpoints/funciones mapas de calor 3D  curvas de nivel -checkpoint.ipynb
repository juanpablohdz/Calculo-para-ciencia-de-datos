```python
from matplotlib import cm # para manejar colores
import matplotlib.pyplot as plt
import numpy as np 
    
```

```python
def f(x,y):
    return np.sin(x) + 2*np.cos(y)


if __name__ == '__main__':

    res = 100

    x = np.linspace(-4, 4, num=res)
    y = np.linspace(-4, 4, num=res)

    x, y = np.meshgrid(x,y)#ejes
    z = f(x,y)

    fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
    surf = ax.plot_surface(x,y,z, cmap=cm.cool)# cmap puntos altos y bajos
    fig.colorbar(surf)
    plt.show()
```

# CURVAS DE NIVEL

```python
#se crean niveles  de curvas  para vizualizar desde arriba 
fig2,ax2=plt.subplots()
level_map=np.linspace(np.min(z),np.max(z),num=res)

#counter plat 
cp=ax2.contour(x,y,z,levels=level_map,cmap=cm.cool)#contourf = llena todo
fig2.colorbar(cp)
```
